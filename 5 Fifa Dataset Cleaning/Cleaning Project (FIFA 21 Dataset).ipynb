#### 1. For Name Columns

```python
# create a function to check if a part of the name is contained in the long name

def check_name(row):
    name = row['Name'] # takes the specified row[''] column name from dataframe
    long_name = row['LongName']  # takes the specified row[''] column name from dataframe
    for n in name.split(): # splits first row into individual word strings
        if n.lower() in long_name.lower(): # converts word strings to lower case and checks if any is in 2nd row column
            return True # returns True if found
    return False # returns False if not found

# apply the check_name function to the dataframe
df['Name in Long Name'] = df.apply(check_name, axis=1)
```

#### 3. Height 

```python
# view type of data in height column

df['Height'].unique()
```

```python
# function to put all the heights in a single unit(cm)

def height_to_cm(Height):
    if 'cm' in Height:
        return int(Height.replace('cm','')) # Replaces 'cm' with empty string if found
    else:
        feet, inches = Height.split("\'") # splits "\'"                           
        inches = inches.replace('"', '') # Replaces '"' with empty string if found
        return round((int(feet)*30.48)+(int(inches)*2.54)) # formula to convert both values from feet, inches to cm
```

```python
#apply the height to cm function on Height column

df['Height']=df['Height'].apply(height_to_cm)
```

```python
# confirm changes

df['Height'].unique()
```

```python
# rename the height column

df = df.rename(columns={'Height': 'Height (cm)'})
```

#### 4. Weight

```python
# view type of data in weight column

df['Weight'].unique()
```

```python
# function to put all the weights in a single unit(kg)

def weight_to_kg(Weight):
    if 'kg' in Weight:
        return int(Weight.replace('kg','')) # replaces 'kg' with empty string if found
    else:
        return round(int(Weight.replace('lbs','')) * 0.453592) # replaces 'lbs' with empty string and converts to kg value
```

```python
# apply the weight to cm function on weight column

df['Weight']=df['Weight'].apply(weight_to_kg)
```

```python
# Confirm changes

df['Weight'].unique()
```

```python
# rename the weight column

df = df.rename(columns={'Weight': 'Weight (kg)'})
```

#### 6. Value, Wage & Release Clause

```python
#check all unique currency and suffices

set([(i[0],i[-1]) for i in pd.unique(df[['Value','Wage','Release Clause']].values.ravel())])
```

```python
 #function to make the money numerical(unit is euro(€))

def clean_money(Money):
    Money = Money.replace('€','')
    multiplier = 1
    if Money[-1]=='M':
        multiplier = 1000000 # multiplies Million(M) values by 1,000,000
        Money = Money[:-1] # removes the last character from money ('€', 'K', 'M') and pass it to money still
    elif Money[-1]=='K':
        multiplier = 1000 # multiplies Thousand(K) values by 1,000
        Money = Money[:-1]
    return round(float(Money) * multiplier)
```

```python
# apply the clean money function on each element in the dataframe

df[['Value','Wage','Release Clause']]=df[['Value','Wage','Release Clause']].applymap(clean_money)
```

```python
# confirm the changes

df[['Value','Wage','Release Clause']].head()
```

```python
# Rename columns

df = df.rename(columns={'Value': 'Value (€)',
                         'Wage': 'Wage (€)',
                         'Release Clause': 'Release Clause (€)'})
```

#### 9. Hits

```python
# check the kind of data in Hits

df.Hits.unique()
```

```python
# function to clean Hits column 

def clean_hits(Hits):
    if isinstance(Hits, str): # checks if column contains strings
        if 'K' in Hits: 
            Hits = int(float(Hits.replace('K', '')) * 1000) # replaces k if found and mupltiplies by 1000
        else:
            Hits = int(Hits) # if k is not found converts values to integer
    elif isinstance(Hits, float): # checks if column contains floats
        Hits = int(Hits) # if yes converts to integer
    return Hits
```

```python
# fill nan with 0, and apply the clean_hits function

df['Hits'] = df['Hits'].fillna('0').apply(clean_hits)
```

```python
# check correction made

df['Hits'].unique()
```

##### 11.2 Contract

```python
# view type of data in Contract column

df['Contract'].unique()
```

```python
#function to sort the Contract column, which return three values, contract_start, contract_end, and contract_type

def contract_sorting(Contract):
    if Contract == 'Free': # checks for where contract is equal to free
        contract_type = 'Free'
        contract_start = 'Available'
        contract_end = 'Not on Contract'
    elif 'On Loan' in Contract: # checks for where contract contains 'on loan'
        contract_type = 'On Loan'
        contract_start = 'Currently On Loan'
        contract_end = Contract.replace('On Loan', '')
    else:
        contract_type = 'On Contract'
        contract_start, contract_end = Contract.split(' ~ ') # splits values where contract contains '~'

    return contract_start, contract_end,contract_type

```

```python
# Create necessary columns and apply the funtion using the Contract column

df[['Contract Start','Contract End', 'Contract Type']]=df.Contract.apply(contract_sorting).apply(pd.Series)
```
